Data Splitting

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

Standard Scaling

```python
#scaling the amount and time columns
from sklearn.preprocessing import StandardScaler
scaler_amount=StandardScaler()
scaler_time=StandardScaler()
df['Amount_scaled']=scaler_time.fit_transform(df['Amount'].values.reshape(-1,1))
df['Time_scaled']=scaler_amount.fit_transform(df['Time'].values.reshape(-1,1))
                                              
                                            
#dropping the initial columns
df.drop(['Amount','Time'],axis=1)
```

Model Training

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
x_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
#Using logistic regression to train the model
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
#Training the Model
y_pred=lr.predict(X_test)
```

Model Evaluation

```python
from sklearn.metrics import classification_report,f1_score
classification_report(y_test,y_pred)
print(classification_report(y_test,y_pred))
print('f1_Score:',f1_score(y_test,y_pred))
print('f1_Score:',f1_score(y_test,y_pred))
```

```python
#using random forest for the model
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
##Training the Model
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,f1_score
classification_report(y_test,y_pred)
print(classification_report(y_test,y_pred))
print('f1_Score:',f1_score(y_test,y_pred))
```

Interpretation Summary: Fraud Detection Model Performance
In our fraud detection task, we evaluated the performance of two machine learning models. The goal was to identify fraudulent transactions (class 1) while maintaining accuracy in detecting non-fraudulent ones (class 0). Below is a summary of the results and key insights from both models:

Model 1: Initial Model Performance (Before RandomForestClassifier)
Non-Fraudulent Transactions (Class 0):

The model performed perfectly in identifying non-fraudulent transactions, with 100% precision and 100% recall. This means that every transaction predicted as non-fraud was correct, and all actual non-fraud transactions were detected.
Fraudulent Transactions (Class 1):

The model had a 75% precision for fraud detection, meaning that 75% of the transactions predicted as fraudulent were correct.
It had a 78% recall, meaning that 78% of actual fraud cases were correctly identified, but 22% were missed.
F1-Score: The balance between precision and recall was 0.76, indicating that while the model is good at detecting fraud, there are some false positives and false negatives.
Overall Accuracy: The overall model accuracy was close to 100%, primarily driven by the large number of non-fraudulent transactions. However, this can be misleading in imbalanced datasets like this, where non-fraud cases dominate.

Model 2: RandomForestClassifier Performance
Non-Fraudulent Transactions (Class 0):

Similar to the first model, this model maintained 100% precision and 100% recall for non-fraudulent transactions. It reliably detects all non-fraudulent transactions without any mistakes.
Fraudulent Transactions (Class 1):

The precision for fraud detection improved to 94%, meaning that 94% of the transactions predicted as fraudulent were truly fraudulent. This is a significant improvement from the initial model, where only 75% of predicted fraud cases were correct.
The recall slightly decreased to 82%, indicating that 82% of actual fraud cases were correctly detected, while 18% were missed.
F1-Score: The F1-score for detecting fraud increased to 0.87, reflecting a better balance between precision and recall. The model is more confident in its fraud predictions with fewer false positives and fewer missed fraud cases compared to the first model.
Overall Accuracy: This model also achieved near-perfect accuracy (100%), though, as before, this is heavily influenced by the model's strong performance in detecting non-fraudulent transactions, which dominate the dataset.
